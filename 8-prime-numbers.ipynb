```python
# function returning true if n is prime
def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    # just a multiple of another prime number
    for i in range(2, int(n/2)):
        if n % i == 0:
            return False
        
    return True



# 136 - Function which returns the list of the prime factors of a number
def prime_factors(number: int) -> list[int]:
    factors = []
    
    for i in range(2, number):
        if (number % i == 0) and is_prime(i):
            factors.append(i)
    
    return factors


def is_quad_prime(n: int) -> bool:
    for i in range(2, int(n/2)):
        if (n % i == 0):
            return False
    
    if ((n - 1) / 4) == 1:
        return True

    return False


# function which finds the 'quad-prime-factors' of a number
# quad prime factors are (in this case) are of the form: 4*n + 1
def quad_prime_factors(n: int) -> list[int]:
    pass


# implementing the sieve of eratosthenes
def sieve(n: int) -> list[int]:
    # list of primes
    primes = []
    # numbers to be sieved
    numbers = [i for i in range(2, n)]
    # starting sieving
    c = 2
    while c * c < n:
        for k in range(c, n, c):
            # sieving out k
            if k in numbers:
                numbers.remove(k)
        # c is prime
        primes.append(c)
        c = numbers[0]
        
    return primes + numbers
```

```python
# 141 - To 'proove' that there is an infinite amount of prime numbers (endless set of primes called P)
def prime_product_up_to_n(n: int):
    p = 1
    for i in range(2, n):
        if is_prime(n):
            p *= i
            
    p += 1
    
    return is_prime(p)
```

```python
# Implementing miller-rabin-test for identifying primes
def miller_rabin(p):
    d = p - 1
    r = 0
    while d % 2 == 0:
        d //= 2
        r += 1
    
    a = random.randrange(2, p - 1)
    # x = (a ** d) % p
    x = pow(a, d, p)
    # a**d mod p == 1?
    if x == 1 or x == p - 1:
        return True
    
    while r > 1:
        x = (x * x) % p
        
        if x == 1:
            return False
        elif x == (p - 1):
            return True

        r -= 1

    return False
```

```python
# 155
def miller_rabin_freq(p, n):
    # counter
    c = 0
    # 157, comparing the ** operator with the pow(base, power, mod[optional]) function
    start = time.time()
    
    for i in range(n):
        if miller_rabin(p):
            c += 1
    
    print(time.time() - start)

    return c / n
```
